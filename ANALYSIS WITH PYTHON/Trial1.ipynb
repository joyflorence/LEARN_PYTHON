```python
data.dtypes
```

```python
# Convert object variables to categorical type
object_vars = data.select_dtypes(include='object').columns
data[object_vars]= data[object_vars].astype('category')
```

```python
# Calculate the correlation matrix for numerical variables
numeric_vars = data.select_dtypes(include=['float64', 'int64']).columns
corr_matrix = data[numeric_vars].corr()
corr_matrix
```

```python
# Plot the correlation matrix as a heatmap using seaborn
sns.heatmap(corr_matrix, annot=True, cmap='cool')
```
