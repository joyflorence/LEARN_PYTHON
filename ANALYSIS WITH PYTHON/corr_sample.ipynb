```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Create a sample dataset with variables of different types
data = {
    'Variable1': ['A', 'B', 'C', 'D', 'E'],
    'Variable2': [2.5, 4.7, 6.2, 8.3, 10.1],
    'Variable3': [5, 10, 15, 20, 25]
}

# Create a pandas DataFrame from the dataset
df = pd.DataFrame(data)
df
```

```python
# Convert object variables to categorical type
object_vars = df.select_dtypes(include='object').columns
df[object_vars] = df[object_vars].astype('category')
```

```python
# Calculate the correlation matrix for numerical variables
numeric_vars = df.select_dtypes(include=['float64', 'int64']).columns
corr_matrix = df[numeric_vars].corr()
```

```python

# Plot the correlation matrix as a heatmap using seaborn
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

```python
# Customize the plot
plt.title('Correlation Matrix')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
```

```python


# Display the plot
plt.show()
```
